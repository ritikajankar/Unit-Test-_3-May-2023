### Qstn5. Who are the majority buyers?(genderwie)

```python
x = data['Gender']
y = data['Amount']

plt.xlabel('Gender')
plt.ylabel('Amount')
plt.bar(x,y, color = 'yellow', edgecolor = 'blue', linewidth = 1)
plt.show()
```

### Qstn 6. where do our majority customers fall(age)?

```python
a=df1['Age'].value_counts().head(10).reset_index()
a.columns=['Age','Amount']
plt.figure(figsize=(10,5))
sns.barplot(x=df1['Age'],y=df1['Amount'],palette='Dark2',edgecolor='black')
plt.title('Majority Customers\n',size=20)
plt.xlabel('Age',size=20)
plt.ylabel('Amount',size=20)
plt.yticks(size=12)
plt.xticks(size=12)
plt.show()
```
